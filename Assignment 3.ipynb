```python
import pandas as pd
from pandas.tools.plotting import scatter_matrix
%matplotlib inline


data = pd.read_csv('iris.txt', sep=',', header=0)
```

```python
data.head(n=0)
#this prints out the column names. It is useful later to make slices of the data to know what columns to pull the data from
```

```python
data.tail()
```

```python
data.describe(exclude=None)
#This calculates the summary statistics for the 4 numerical columns. From this data we can see the below information:
#1. The PetalLength has the highest variation
#2. 
```

```python
data.plot.box()
```

```python
data.plot()
#The plot line shows the groupings for each column. You can see the Petal length has the highest
#variability. Sepal width has the lowest variabilitiy.
```

```python
data.plot.hist()
```

```python
data.hist()
#The histogram of each column shows the frequency of each value. The histograms show that looking at the data by the species might give us a 
#better view of each data set. 
#1. PetalLength: The most frenquent value is between 1-2. There is then a break of values and looks to become closer to normal 
#3-7. 
#2. SepalWidth is closered around 3.25. This make sense because it has the lowest standard deviation. 
```

```python
data.groupby('Species').describe()
#By grouping the summary statistics We can see that SepalWidth is the worst characteristic to use to determine which species
#a flower would be by using this statistic. The mean/standard deviation of each species are too close to easily identify what
#species a given SepalWidth would be. 
```

```python
plotting = data.boxplot(by='Species')
#This bloxplot shows the range and mean of each characteristic of the species. Based on this information PetalLength or
#PetalWidth look to be the best options for choosing a single characteristic. 
```

```python
data.groupby('Species').plot.box()
```

```python
data.groupby('Species').plot.line()
```

```python
matrix = scatter_matrix(data, alpha=0.2, figsize=(15, 15), diagonal='hist')
#The scatter matrix shows the correlation between each of the characteristics and each other and themselves. 
#
```

Based on the graphics above, I would choose either PetalLength or PetalWidth to determine which species a specific flower is
Both of these characteristics have good differentiation between the species, especially for Setosa. In the case of this 
species, the maximum PetalWidth is lower than the minimum of the other species, so if I knew the 4 charactersitics of a flower
I would use PetalWidth first. Then I would check PetalLength. Similarly to PetalWidth, there is significant differentiationfor the different species, particularly Setosa. However, Versicolor and Virginica are also differentiated, just not to the same degree.
